```python
#Membuat fungsi untuk menambahkan data ke file CSV yang sudah ada

def automatic_check_price():

    # import Library yang dibutuhkan
    from bs4 import BeautifulSoup
    import pandas as pd
    import requests
    import time
    import datetime
    import csv
    import datetime
    
    # Koneksi Ke Website
    URL = "https://www.tokopedia.com/ismileofficial/apple-iphone-14-pro-garansi-resmi-128gb-256gb-512gb-1tb-128gb-space-black"
    headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/116.0.0.0 Safari/537.36"}

    page = requests.get(URL, headers=headers)

    soup1 = BeautifulSoup(page.content, "html.parser")
    soup2 = BeautifulSoup(soup1.prettify(), "html.parser")

    #Mengambil Judul Produk
    title = soup2.find(class_="css-1os9jjn").get_text()
    title = title.strip()

    #Mengambil Harga Produk
    price = soup2.find(class_="price").get_text()
    price = price.strip()[2:]

    #Today
    today = datetime.date.today()
    
    header = ['Judul',"Harga","WaktuAmbil"]
    data = [title, price, today]
    
    with open("TokpedWebScraper.csv","a+",newline = "", encoding = 'UTF8') as f:
        writer = csv.writer(f)
        writer.writerow(data)
        
    # Menggunakan fungsi untuk mengirim email jika harga Iphone 14 < Rp. 12.000.000
    if price <= 12000000:
        send_me_notification("Harga Iphone Turun!!!", "Buruan Checkout!")
```

```python
#Mengambil data tiap 3 detik selama 30 detik
start_time = time.time()
 
while(time.time() - start_time <30):
    automatic_check_price()
    time.sleep(3)
```

```python
#Fungsi untuk mengirim notifikasi jika harga Iphone 14 < dari nilai tertentu
import smtplib
from email.message import EmailMessage

def send_me_notification(subject, body):
    # Email saya
    my_email = "mayerdsiahaan@gmail.com"
    my_password = "************"
    
    # Membuat pesan email
    msg = EmailMessage()
    msg.set_content(body)
    msg["Subject"] = subject
    msg["From"] = my_email
    msg["To"] = my_email
    
    # Menggunakan SMTP server dari Gmail
    server = smtplib.SMTP("smtp.gmail.com", 587)
    server.starttls()  
    server.login(my_email, my_password)
    server.send_message(msg)
    server.quit()
```
